```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv('final_data.csv')
```

```python
df.describe()
```

```python
# Number of players each year

df['year'] = pd.to_datetime(df['posted']).dt.year
players_per_year = df.groupby('year')['user_id'].nunique()
plt.figure(figsize=(10, 6))
players_per_year.plot(kind='bar')
plt.title('Number of Players per Year')
plt.xlabel('Year')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
#Distribution of Genres

genre_counts = df['genres'].value_counts()
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar')
plt.title('Distribution of Game Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Playtime per year by genre

genre_playtime = df.groupby(['year', 'genres'])['playtime_forever'].sum().unstack()

plt.figure(figsize=(12, 6))
genre_playtime.plot(kind='bar', stacked=True)
plt.title('Playtime by Genre per Year')
plt.xlabel('Year')
plt.ylabel('Total Playtime (hours)')
plt.xticks(rotation=0)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Sentiment Analysis Distribution

plt.figure(figsize=(8, 6))
df['sentiment_analysis'].plot(kind='hist', bins=20)
plt.title('Distribution of Sentiment Analysis Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```
